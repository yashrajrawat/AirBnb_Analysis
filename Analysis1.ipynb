```python
def rev_hist(df,bins,l):
    plt.hist(df,bins,color="#76528BFF",label='Review Count')
    plt.legend()
    plt.xlim(l)
    plt.xlabel('Count')
    plt.ylabel('Users')
    plt.show()
```

```python
def ratings_overall(x,p):
    (airbnb[(x>=p[0]) & (x<=p[1])]['overall_satisfaction']).plot(kind='hist',color='#DF6589FF',label='Ratings')
    plt.grid(alpha=0.4,color="#3C1053FF")
    plt.title(f'House Prices Satisfaction( Price $:{p[0]}- {p[1]})')
    plt.legend()
    plt.show()
```

```python
ratings_overall(airbnb['price'],[1,100])
```

```python
ratings_overall(airbnb['price'],[1,50])
```

```python
ratings_overall(airbnb['price'],[200,6000])
```

```python
ratings_overall(airbnb['price'],[300,6000])
```
